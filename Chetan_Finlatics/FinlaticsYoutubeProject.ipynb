# Question 8:

```python
print("NaN values in created_year column:", df['created_year'].isna().sum())
df['created_year'] = df['created_year'].fillna(2020).astype(int)  # Fill NaN with a default value
df['created_datetime'] = pd.to_datetime(df['created_year'].astype(str) + '-' + df['created_month'] + '-01')

monthly_trend = df.groupby('created_datetime').agg({
    'subscribers_for_last_30_days': 'sum'
}).reset_index()

plt.figure(figsize=(10, 6))
plt.plot(monthly_trend['created_datetime'], monthly_trend['subscribers_for_last_30_days'], marker='o', linestyle='-')

plt.title('Overall Trend in Subscribers Gained in the Last 30 Days Across All Channels')
plt.xlabel('Date')
plt.ylabel('Total Subscribers Gained in Last 30 Days')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from datetime import datetime
# Map month abbreviations to numbers
month_mapping = {
    'Jan': 1,
    'Feb': 2,
    'Mar': 3,
    'Apr': 4,
    'May': 5,
    'Jun': 6,
    'Jul': 7,
    'Aug': 8,
    'Sep': 9,
    'Oct': 10,
    'Nov': 11,
    'Dec': 12
}

today=datetime.today()
df['m_num'] = df['created_month'].map(month_mapping)
df['num_months']=((today.year-df['created_year'])*12)+(today.month-df['m_num'])
df['Average number of subscribers gained']=df['subscribers']/df['num_months']
print(f"Average subscriber gained per month = {df['Average number of subscribers gained'].mean()}")
df[['Youtuber','Average number of subscribers gained']].head(10)
                            
```
